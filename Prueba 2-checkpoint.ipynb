```python
n_datos = int(len(NLG['ROIs'])-0.19*len(NLG['ROIs']))
test_dataset = NLG[n_datos:134114]
test_dataset.reset_index(drop = True)
training_dataset = NLG[0:(n_datos-1)]
test_dataset.reset_index(drop = True)
```

#### Procesado de los datos

```python
labels_dataset = pd.DataFrame(training_dataset['ROIs'])
training_dataset = training_dataset.drop('ROIs', 1)
training_dataset
```

```python
training_dataset.reset_index(drop = True)
labels_dataset.reset_index(drop = True)
```

Si se hacen comparaciones entre otras etiquetas se han de cambiar los strings N y LG por las que sean de interes.

```python
labels_dataset['Code'] = 0

i=0
for x in labels_dataset['ROIs']:
    if x == 'N': labels_dataset['Code'].array[i]=0
    if x == 'LG': labels_dataset['Code'].array[i]=1
    i = i + 1
            
labels_dataset.reset_index(drop = True)
```

```python
dfx = np.array(training_dataset)
dfx.shape
```

```python
dfy = np.array(labels_dataset['Code'])
dfy.shape
```

```python
scaler = MinMaxScaler(feature_range=(0,1))
scaled_dfx = scaler.fit_transform(dfx)
scaled_dfx
```

#### Caracterización del modelo predictivo

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Activation
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.metrics import categorical_crossentropy
```

```python
model = Sequential([
    Dense(units = 64, input_shape = (1236,), activation = 'relu'),
    Dense(units = 128, activation = 'relu'),
    Dense(units = 2, activation = 'softmax')

])
```

```python
model.compile(optimizer = Adam(learning_rate = 0.01), loss = 'sparse_categorical_crossentropy', metrics = ['accuracy'])
```

```python
model.fit(x = scaled_dfx, y = dfy,validation_split = 0.05,shuffle = True, batch_size = 250, epochs= 10, verbose =  2)
```

#### Procesado de los datos de prueba

```python
ol = pd.DataFrame(test_dataset['ROIs'])

ol['Code'] = 0

i=0
for x in ol['ROIs']:
    if x == 'N': ol['Code'].array[i]=0
    if x == 'LG': ol['Code'].array[i]=1
    i = i + 1
            
labels_dataset.reset_index(drop = True)
td = test_dataset.drop('ROIs', 1)
ol.reset_index(drop = True)
td.reset_index(drop = True)
```

```python
n1 = 0
n2 = 0


for x in testy1:
    if x == 0: n1 = n1 + 1
    if x == 1: n2 = n2 + 1
 

n1,n2
```
